```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("C:\\Users\\NITJ\\Desktop\\DMMidTerm_TeamId_37\\Data\\8.3_Vehicles_in_use_as_on_31st_December_1969_and_1970_in_certain_foreign_countries.csv")
#Removing rows and columns which have excess NA's(Data Reduction)
df=df.dropna(thresh=len(df)-15,axis=1)
df=df.dropna(axis=0, how='any', thresh=2, subset=None, inplace=False)
#Filling NA with suitable central tendency(Data cleaning)
df=df.fillna(df.mean(),inplace=False)
df
```

```python
#Combining group of columns into one one cloumn(Data Aggregation)
df['Four-plus wheels - Total Cols. (3) to (6) (Col. 7)']=df['Four-plus wheels - Cars (Col. 3)']+df['Four-plus wheels - Buses & Coaches (Col. 4)']+df['Four-plus wheels - Goods Vehicles (Col. 5)']+df['Four-plus wheels - Tractors (excluding agricultural tractors) (Col. 6)']+df['Trailers (Col. 11)']
del df['Trailers (Col. 11)']
del df['Four-plus wheels - Cars (Col. 3)']
del df['Four-plus wheels - Buses & Coaches (Col. 4)']
del df['Four-plus wheels - Goods Vehicles (Col. 5)']
del df['Four-plus wheels - Tractors (excluding agricultural tractors) (Col. 6)']
df=df.rename(columns = {"Four-plus wheels - Total Cols. (3) to (6) (Col. 7)": "Four plus wheels"})
df
```

```python
#Combining group of columns into one one cloumn(Data Aggregation)
df['Two wheels - Total cols. (8) & (9) (Col. 10)']=df['Two wheels - Motor Cycles & Scooters (Col. 8)']+df['Two wheels - Mopeds (Col. 9)']
del df['Two wheels - Mopeds (Col. 9)']
del df['Two wheels - Motor Cycles & Scooters (Col. 8)']
df=df.rename(columns = {"Two wheels - Total cols. (8) & (9) (Col. 10)": "Two plus wheels"})
df
```

```python
#Scatter plot of country-wise two wheelers in the year 1969 for the visualisation
df1=df[df['Year (Col. 2)']==1969]
plt.scatter(df1['Two plus wheels'], df1['Country (Col. 1)'])
plt.show() 
```

```python
#Scatter plot of country-wise two wheelers in the year 1970
df1=df[df['Year (Col. 2)']==1970]
plt.scatter(df1['Two plus wheels'], df1['Country (Col. 1)'])
plt.show() 
```

```python
#Scatter plot of country-wise four wheelers in the year 1969
df1=df[df['Year (Col. 2)']==1969]
plt.scatter(df1['Four plus wheels'], df1['Country (Col. 1)'])
plt.show() 
```

```python
#Scatter plot of country-wise four wheelers in the year 1970
df1=df[df['Year (Col. 2)']==1970]
plt.scatter(df1['Four plus wheels'], df1['Country (Col. 1)'])
plt.show() 
```
