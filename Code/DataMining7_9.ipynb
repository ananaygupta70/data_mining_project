```python
import pandas as pd
df=pd.read_csv("C:\\Users\\NITJ\\Desktop\\DMMidTerm_TeamId_37\\Data\\7.9_State_wise_Rates_of_Accidents_by_Class_of_Motor_Vehicles_on_Road_1971_per_1000_vehicles.csv")
#Replacing all non-numerical values with NA(Data Cleaning)
df=df.replace(to_replace ="@",value ="NaN")
df=df.replace(to_replace ="*",value ="NaN")
#Removing all rows and columns having excess NA
df=df.dropna(thresh=len(df)-3,axis=1)
df=df.dropna(axis=0, how='any', thresh=8, subset=None, inplace=False)
#Filling NA with suitable central tendencies
df=df.fillna(df.mean(),inplace=False)
df
```

```python
#Creating bar chart for visualization
df1=df
del df1['All types of vehicles (Col. 3)']
df1.set_index("State/ Union Territory (Col. 2)",inplace=True)
import matplotlib.pyplot as plt
df1.plot.bar(stacked=False,figsize=(40,20))
plt.show()
```
