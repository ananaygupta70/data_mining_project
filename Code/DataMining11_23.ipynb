```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("C:\\dataMining\\11.23_Seating_capacity_and_Kilometrage_in_different_undertakings_Service_wise.csv")
#Replacing the non-numerical data with NA(Data Cleaning)
df=df.replace(to_replace ="Seats",value ="NaN")
df=df.replace(to_replace ="Otls.",value ="NaN")
#Removing rows and columns with excess NA(Data Reduction)
df=df.dropna(thresh=len(df)-12,axis=1)
df=df.dropna(axis=0, how='any', thresh=7, subset=None, inplace=False)
#Replacing the NA values with median i.e. suitable central tendency(data cleaning)
df=df.fillna(df.median(),inplace=False)
df
```

```python
#Performing Data Aggregation(Combining fields of same category)
atot1=df['Fleet operated (Col. 5)'].iloc[0:12].sum()
atot2=df['Fleet operated (Col. 5)'].iloc[12:30].sum()
atot3=df['Fleet operated (Col. 5)'].iloc[30:36].sum()
atot4=df['Fleet operated (Col. 5)'].iloc[36:44].sum()
atot5=df['Fleet operated (Col. 5)'].iloc[44:50].sum()
btot1=df['Seating capacity per bus (Col. 6)'].iloc[0:12].sum()
btot2=df['Seating capacity per bus (Col. 6)'].iloc[12:30].sum()
btot3=df['Seating capacity per bus (Col. 6)'].iloc[30:36].sum()
btot4=df['Seating capacity per bus (Col. 6)'].iloc[36:44].sum()
btot5=df['Seating capacity per bus (Col. 6)'].iloc[44:50].sum()
ctot1=df['Total seating capacity of fleet operated (Col. 7)'].iloc[0:12].sum()
ctot2=df['Total seating capacity of fleet operated (Col. 7)'].iloc[12:30].sum()
ctot3=df['Total seating capacity of fleet operated (Col. 7)'].iloc[30:36].sum()
ctot4=df['Total seating capacity of fleet operated (Col. 7)'].iloc[36:44].sum()
ctot5=df['Total seating capacity of fleet operated (Col. 7)'].iloc[44:50].sum()
dtot1=df['Effective kilometres (in lakhs) (Col. 8)'].iloc[0:12].sum()
dtot2=df['Effective kilometres (in lakhs) (Col. 8)'].iloc[12:30].sum()
dtot3=df['Effective kilometres (in lakhs) (Col. 8)'].iloc[30:36].sum()
dtot4=df['Effective kilometres (in lakhs) (Col. 8)'].iloc[36:44].sum()
dtot5=df['Effective kilometres (in lakhs) (Col. 8)'].iloc[44:50].sum()
etot1=df['Gross kilometres (In lakhs) (Col. 9)'].iloc[0:12].sum()
etot2=df['Gross kilometres (In lakhs) (Col. 9)'].iloc[12:30].sum()
etot3=df['Gross kilometres (In lakhs) (Col. 9)'].iloc[30:36].sum()
etot4=df['Gross kilometres (In lakhs) (Col. 9)'].iloc[36:44].sum()
etot5=df['Gross kilometres (In lakhs) (Col. 9)'].iloc[44:50].sum()
data={
    'Sub-category Service-wise':["City Services","District-Plain Services","District-Hill Services","Goods Services	","Taxi Services"],    
    'Fleet operated Service-wise':[atot1,atot2,atot3,atot4,atot5],
    'Seating capacity per bus Service-wise':[btot1,btot2,btot3,btot4,btot5],
    'Total seating capacity of fleet operated Service-wise':[ctot1,ctot2,ctot3,ctot4,ctot5],
    'Effective kilometres Service-wise':[dtot1,dtot2,dtot3,dtot4,dtot5],
    'Gross kilometres Service-wise':[etot1,etot2,etot3,etot4,etot5]
        }
dtf=pd.DataFrame(data,columns=["Sub-category Service-wise","Fleet operated Service-wise","Seating capacity per bus Service-wise","Total seating capacity of fleet operated Service-wise","Effective kilometres Service-wise","Gross kilometres Service-wise"])
dtf.set_index("Sub-category Service-wise",inplace=True)
dtf
```

```python
#Data Normalisation for proper data visualisation
dtf['Total seating capacity of fleet operated Service-wise']=dtf['Total seating capacity of fleet operated Service-wise']/1000
dtf['Fleet operated Service-wise']=dtf['Fleet operated Service-wise']/10
dtf['Effective kilometres Service-wise']=dtf['Effective kilometres Service-wise']/10
dtf['Gross kilometres Service-wise']=dtf['Gross kilometres Service-wise']/10
#Plotting the bar graph of different categories
dtf.plot.bar(stacked=False,figsize=(20,20))
plt.show()
```
