```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("C:\\Users\\NITJ\\Desktop\\DMMidTerm_TeamId_37\\Data\\8.1_Production_of_Motor_Vehicles_in_certain_foreign_countries.csv",header= 0,encoding= 'unicode_escape')
#dropping all rows and columns with excessive NA's(Data reduction)
df=df.dropna(thresh=len(df)-16,axis=1)
df=df.dropna(axis=0, how='any',thresh=3,subset=None,inplace=False)
#Filling NA with the mean(Data Cleaning)
df=df.fillna(df.mean(),inplace=False)
df
```

```python
#Combing group of attributes into one attribute(Aggregation)
df['Goods Vehicles (Col. 5)']=df['Goods Vehicles (Col. 5)']+df['Private Cars (Col. 3)']+df['Buses & Coaches (Col. 4)']
df['Motor Cycles & Scooters (Col. 6)']=df['Motor Cycles & Scooters (Col. 6)']+df['Mopeds (Col. 7)']
del df['Note of Goods Vehicles']
del df['Private Cars (Col. 3)']
del df['Buses & Coaches (Col. 4)']
del df['Mopeds (Col. 7)']
df=df.rename(columns = {"Goods Vehicles (Col. 5)":"Four plus wheels"})
df=df.rename(columns = {"Motor Cycles & Scooters (Col. 6)":"Two wheels"})
df
```

```python
#Scatter plot of country-wise two wheelers in the year 1969
df1=df[df['Year (Col. 2)']==1969]
plt.scatter(df1['Two wheels'], df1['Country (Col. 1)'])
plt.show() 
```

```python
#Scatter plot of country-wise two wheelers in the year 1970
df2=df[df['Year (Col. 2)']==1970]
plt.scatter(df2['Motor Cycles & Scooters (Col. 6)'], df2['Country (Col. 1)'])
plt.show() 
```

```python
#Scatter plot of country-wise four wheelers in the year 1969
plt.scatter(df1['Four plus wheels'], df1['Country (Col. 1)'])
plt.show() 
```

```python
#Scatter plot of country-wise four wheelers in the year 1970
plt.scatter(df2['Four plus wheels'], df2['Country (Col. 1)'])
plt.show() 
```
