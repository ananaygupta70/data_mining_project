```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("C:\\Users\\NITJ\\Desktop\\DMMidTerm_TeamId_37\\Data\\11.5_Total_Costs_of_Operation_on_Public_Sector_Road_Transport_Undertakings_1960_61_to_1970_71.csv")
#Replacing all the non-numerical values with NA(Data Cleaning)
df=df.replace(to_replace ="@",value ="NaN")
df=df.replace(to_replace ="*",value ="NaN")
#Removing rows and columns with NA(Data Reduction)
df=df.dropna(thresh=len(df)-5,axis=1)
df=df.dropna(axis=0, how='any', thresh=6, subset=None, inplace=False)
#Taking row-wise mean and filling NA values(Data Cleaning)
df=df.apply(lambda row: row.fillna(row[['1960-61 (Col.3)','1965-66 (Col.4)','1967-68 (Col.5)','1968-69 (Col.6)','1969-70 (Col.7)','1970-71 (Col.8)']].mean()), axis=1)
#Removing the column because of so much irregularities(Removing highly irregular data)
del df['1960-61 (Col.3)']
df
```

```python
#Performing data aggregation(Combing fields of the same category)
atot1=df['1965-66 (Col.4)'].iloc[0:4].sum()
atot2=df['1965-66 (Col.4)'].iloc[4:11].sum()
atot3=df['1965-66 (Col.4)'].iloc[11:14].sum()
atot4=df['1965-66 (Col.4)'].iloc[14:19].sum()
atot5=df['1965-66 (Col.4)'].iloc[19:20].sum()
btot1=df['1967-68 (Col.5)'].iloc[0:4].sum()
btot2=df['1967-68 (Col.5)'].iloc[4:11].sum()
btot3=df['1967-68 (Col.5)'].iloc[11:14].sum()
btot4=df['1967-68 (Col.5)'].iloc[14:19].sum()
btot5=df['1967-68 (Col.5)'].iloc[19:20].sum()
ctot1=df['1968-69 (Col.6)'].iloc[0:4].sum()
ctot2=df['1968-69 (Col.6)'].iloc[4:11].sum()
ctot3=df['1968-69 (Col.6)'].iloc[11:14].sum()
ctot4=df['1968-69 (Col.6)'].iloc[14:19].sum()
ctot5=df['1968-69 (Col.6)'].iloc[19:20].sum()
dtot1=df['1969-70 (Col.7)'].iloc[0:4].sum()
dtot2=df['1969-70 (Col.7)'].iloc[4:11].sum()
dtot3=df['1969-70 (Col.7)'].iloc[11:14].sum()
dtot4=df['1969-70 (Col.7)'].iloc[14:19].sum()
dtot5=df['1969-70 (Col.7)'].iloc[19:20].sum()
etot1=df['1970-71 (Col.8)'].iloc[0:4].sum()
etot2=df['1970-71 (Col.8)'].iloc[4:11].sum()
etot3=df['1970-71 (Col.8)'].iloc[11:14].sum()
etot4=df['1970-71 (Col.8)'].iloc[14:19].sum()
etot5=df['1970-71 (Col.8)'].iloc[19:20].sum()
data={
    'Categories of expenditure':["Cost of personnel employed","Cost of materials consumed","Depreciation on assets","Overhead costs","Interest on borrowed capital"],    
    '1965-66':[atot1,atot2,atot3,atot4,atot5],
    '1967-68':[btot1,btot2,btot3,btot4,btot5],
    '1968-69':[ctot1,ctot2,ctot3,ctot4,ctot5],
    '1969-70':[dtot1,dtot2,dtot3,dtot4,dtot5],
    '1970-71':[etot1,etot2,etot3,etot4,etot5]
        }
dtf=pd.DataFrame(data,columns=["Categories of expenditure","1965-66","1967-68","1968-69","1969-70","1970-71"])
dtf.set_index("Categories of expenditure",inplace=True)
dtf
```

```python
#Horizontal Bar Graph of different categories year-wise
dtf.plot.barh(figsize=(20,10))
plt.show()
```
