```python
import pandas as pd
df=pd.read_csv("C:\\Users\\NITJ\\Desktop\\DMMidTerm_TeamId_37\\Data\\7.3_Motor_Vehicle_Accidents_during_1971_Month_wise.csv")
#Replacing all the non-numerical values with NA(Data Cleaning)
df=df.replace(to_replace ="@",value ="NaN")
df=df.replace(to_replace ="*",value ="NaN")
#Removing rows and columns with excess NA's
df=df.dropna(thresh=len(df)-5,axis=1)
df=df.dropna(axis=0, how='any', thresh=11, subset=None, inplace=False)
#Filling all the NA with the mean
df=df.fillna(df.mean(),inplace=False)
df
```
