```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("C:\\dataMining\\7.20_Motor_Vehicle_Accidents_during_1970_by_Condition_of_Motor_Vehicles_Involved.csv")
#Replacing all the non-numerical values with NA(data cleaning)
df=df.replace(to_replace ="@",value ="NaN")
df=df.replace(to_replace ="*",value ="NaN")
#Removing the rows and columns having excessive NA's(Data Reduction)
df=df.dropna(thresh=len(df)-7,axis=1)
df=df.dropna(axis=0, how='any', thresh=4, subset=None, inplace=False)
#Filling all the NA with mean(Data Cleaning)
df=df.fillna(df.mean(),inplace=False)
df
```

```python
#Combining group of attributes into a single attribute
df['Other serious mechanical defects (Col. 9)']=df['Other serious mechanical defects (Col. 9)']+df['Defective brakes (Col. 5)']+df['Defective Steering (Col. 6)']
del df['Defective Steering (Col. 6)']
del df['Defective brakes (Col. 5)']
df=df.rename(columns = {"Other serious mechanical defects (Col. 9)":"Mechanical Defects"})
df
```

```python
del df['Total (Col. 3)']
#Changing the index
df.set_index('State/ Union Territory (Col. 2)',inplace=True)
#Plotting the stacked bar graph
df.plot.bar(stacked=True,figsize=(30,40))
plt.show()
```
