```python
#code section for datacleaning
import pandas as pd #importing necessary libraries
import matplotlib.pyplot as plt
df=pd.read_csv("C:\\Users\\NITJ\\Desktop\\DMMidTerm_TeamId_37\\Data\\11.29_Contribution_of_Individual_Public_Sector_Road_Transport_Undertakings_to_National_Income.csv") #selecting data frame to perform data preprocessing
df
```

```python
#code section for merging data of different rows and columns and removing redundant rows and columns
atot1=df['Total expenditure on personnel (Col. 5)'].iloc[0:7].sum() #taking summation of necessary values in their respective locations
atot2=df['Total expenditure on personnel (Col. 5)'].iloc[7:18].sum()
atot3=df['Total expenditure on personnel (Col. 5)'].iloc[18:21].sum()
atot4=df['Total expenditure on personnel (Col. 5)'].iloc[21:28].sum()
atot5=df['Total expenditure on personnel (Col. 5)'].iloc[28:29].sum()
btot1=df['Current expenditure (Col. 6)'].iloc[0:7].sum()
btot2=df['Current expenditure (Col. 6)'].iloc[7:18].sum()
btot3=df['Current expenditure (Col. 6)'].iloc[18:21].sum()
btot4=df['Current expenditure (Col. 6)'].iloc[21:28].sum()
btot5=df['Current expenditure (Col. 6)'].iloc[28:29].sum()
ctot1=df['Net profits (Col. 7)'].iloc[0:7].sum()
ctot2=df['Net profits (Col. 7)'].iloc[7:18].sum()
ctot3=df['Net profits (Col. 7)'].iloc[18:21].sum()
ctot4=df['Net profits (Col. 7)'].iloc[21:28].sum()
ctot5=df['Net profits (Col. 7)'].iloc[28:29].sum()
dtot1=df['Total Contribution (Col. 8)'].iloc[0:7].sum()
dtot2=df['Total Contribution (Col. 8)'].iloc[7:18].sum()
dtot3=df['Total Contribution (Col. 8)'].iloc[18:21].sum()
dtot4=df['Total Contribution (Col. 8)'].iloc[21:28].sum()
dtot5=df['Total Contribution (Col. 8)'].iloc[28:29].sum()
data={
    'Sub-category Service-wise':["City Services"," District Services","City and District Services","Passenger-cum-Goods Transport","Goods Transport"],    
    'Total expenditure on personnel Service-wise':[atot1,atot2,atot3,atot4,atot5],
    'Current expenditure Service-wise':[btot1,btot2,btot3,btot4,btot5],
    'Net profits Service-wise':[ctot1,ctot2,ctot3,ctot4,ctot5],
    'Total Contribution Service-wise':[dtot1,dtot2,dtot3,dtot4,dtot5]
        } #adding data to a new dataframe with previously merged values
dtf=pd.DataFrame(data,columns=["Sub-category Service-wise","Total expenditure on personnel Service-wise","Current expenditure Service-wise","Net profits Service-wise","Total Contribution Service-wise"])
dtf.set_index("Sub-category Service-wise",inplace=True) #setting index to Sub-category Service-wise
dtf
```
