```python
#importing required libraries
import pandas as pd
import matplotlib.pyplot as plt

#reading data from storage location
df=pd.read_csv("C:\\Users\\NITJ\Desktop\\DMMidTerm_TeamId_37\\Data\\10.9_Total_No_of_Accidents_and_Accidents_per_lakh_gross_kilometres.csv")
df1=pd.read_csv("C:\\Users\\NITJ\\Desktop\\DMMidTerm_TeamId_37\\Data\\10.8_Total_No_of_Breakdowns_and_No_of_breakdowns_per_10000_Gross_Kms.csv")
df
```

```python
#Deleting duplicate fields for merging
del df1['Category (Col. 1)']
del df1['Name of the Undertaking (Col. 2)']

#Merging two data
df=df.join(df1)

#replacing all non numerical values(garbage) to null values(Data Cleaning)
df=df.replace(to_replace ="*", value ="NaN")

#dropping columns and rows respectively with excess null values(Data Reduction)
df=df.dropna(thresh=len(df)-6,axis=1)
df=df.dropna(axis=0, how='any',thresh=10,subset=None,inplace=False)

#filling null values in particular column with suitable central tendency measure
df=df.fillna(df.iloc[[35,42]].median(),inplace=False)
df
```

```python

#making data equivalent with multiplying with suitable value (normalization)
df['No. of breakdowns per 10000 G. Kms. During the quarter ending - June 1970 (Col. 7)']=df['No. of breakdowns per 10000 G. Kms. During the quarter ending - June 1970 (Col. 7)']*10
df['No. of breakdowns per 10000 G. Kms. During the quarter ending - Sept. 1970 (Col. 8)']=df['No. of breakdowns per 10000 G. Kms. During the quarter ending - Sept. 1970 (Col. 8)']*10

#renaming data for proper intference
df=df.rename(columns={"No. of breakdowns per 10000 G. Kms. During the quarter ending - June 1970 (Col. 7)":"No. of breakdowns per lakh G. Kms. During the quarter ending - June 1970 (Col. 7)"})
df=df.rename(columns={"No. of breakdowns per 10000 G. Kms. During the quarter ending - Dec. 1970 (Col. 9)":"No. of breakdowns per lakh G. Kms. During the quarter ending - Dec. 1970 (Col. 9)"})
df=df.rename(columns={"No. of breakdowns per 10000 G. Kms. During the quarter ending - Sept. 1970 (Col. 8)":"No. of breakdowns per lakh G. Kms. During the quarter ending - Sept. 1970 (Col. 8)"})
df=df.rename(columns={"No. of breakdowns per 10000 G. Kms. During the quarter ending - March 1971 (Col. 10)":"No. of breakdowns per lakh G. Kms. During the quarter ending - March 1971 (Col. 10)"})

#Merging two columns to give a broader sense (Data Reduction)
df['Total Number of Accidents during the quarter ending - Dec. 1970 (Col. 5)']=df['Total Number of Accidents during the quarter ending - Dec. 1970 (Col. 5)']+df['Total Number of Accidents during the quarter ending - March 1971 (Col. 6)']
df['Accidents per lakh gross kilometres during the quarter ending - Dec. 1970 (Col. 9)']=df['Accidents per lakh gross kilometres during the quarter ending - Dec. 1970 (Col. 9)']+df['Accidents per lakh gross kilometres during the quarter ending - March 1971 (Col. 10)']
df['No. of breakdowns during the quarter ending - June 1970 (Col. 3)']=df['No. of breakdowns during the quarter ending - June 1970 (Col. 3)']+df['No. of breakdowns during the quarter ending - Sept. 1970 (Col. 4)']+df['No. of breakdowns during the quarter ending - Dec. 1970 (Col. 5)']+df['No. of breakdowns during the quarter ending - March 1971 (Col. 6)']
df['No. of breakdowns per lakh G. Kms. During the quarter ending - June 1970 (Col. 7)']=df['No. of breakdowns per lakh G. Kms. During the quarter ending - June 1970 (Col. 7)']+df['No. of breakdowns per lakh G. Kms. During the quarter ending - Sept. 1970 (Col. 8)']+df['No. of breakdowns per lakh G. Kms. During the quarter ending - Dec. 1970 (Col. 9)']+df['No. of breakdowns per lakh G. Kms. During the quarter ending - March 1971 (Col. 10)']

#Deleting nonessential data (Data Reduction)
del df['Total Number of Accidents during the quarter ending - March 1971 (Col. 6)']
del df['Accidents per lakh gross kilometres during the quarter ending - March 1971 (Col. 10)']
del df['No. of breakdowns during the quarter ending - March 1971 (Col. 6)']
del df['No. of breakdowns during the quarter ending - Dec. 1970 (Col. 5)']
del df['No. of breakdowns during the quarter ending - Sept. 1970 (Col. 4)']
del df['No. of breakdowns per lakh G. Kms. During the quarter ending - Sept. 1970 (Col. 8)']
del df['No. of breakdowns per lakh G. Kms. During the quarter ending - Dec. 1970 (Col. 9)']
del df['No. of breakdowns per lakh G. Kms. During the quarter ending - March 1971 (Col. 10)']

#Renaming data for proper inference
df=df.rename(columns={"Total Number of Accidents during the quarter ending - Dec. 1970 (Col. 5)":"Total Number of Accidents during the half year ending - march. 1971 (Col. 3)"})
df=df.rename(columns={"Accidents per lakh gross kilometres during the quarter ending - Dec. 1970 (Col. 9)":"Accidents per lakh gross kilometres during the half year ending - March. 1971 (Col. 4)"})
df=df.rename(columns={"No. of breakdowns during the quarter ending - June 1970 (Col. 3)":"No. of breakdowns during the year ending - March 1971 (Col. 5)"})
df=df.rename(columns={"No. of breakdowns per lakh G. Kms. During the quarter ending - June 1970 (Col. 7)":"No. of breakdowns per lakh G. Kms. During the year ending - March 1971 (Col. 6)"})
df
```

```python

#Summing up data into sub-categories row wise to present a proper view in broader sense
atot1=df['Total Number of Accidents during the half year ending - march. 1971 (Col. 3)'].iloc[0:12].sum()
atot2=df['Total Number of Accidents during the half year ending - march. 1971 (Col. 3)'].iloc[12:30].sum()
atot3=df['Total Number of Accidents during the half year ending - march. 1971 (Col. 3)'].iloc[30:36].sum()
atot4=df['Total Number of Accidents during the half year ending - march. 1971 (Col. 3)'].iloc[36:42].sum()

btot1=df['Accidents per lakh gross kilometres during the half year ending - March. 1971 (Col. 4)'].iloc[0:12].sum()
btot2=df['Accidents per lakh gross kilometres during the half year ending - March. 1971 (Col. 4)'].iloc[12:30].sum()
btot3=df['Accidents per lakh gross kilometres during the half year ending - March. 1971 (Col. 4)'].iloc[30:36].sum()
btot4=df['Accidents per lakh gross kilometres during the half year ending - March. 1971 (Col. 4)'].iloc[36:42].sum()

ctot1=df['No. of breakdowns during the year ending - March 1971 (Col. 5)'].iloc[0:12].sum()
ctot2=df['No. of breakdowns during the year ending - March 1971 (Col. 5)'].iloc[12:30].sum()
ctot3=df['No. of breakdowns during the year ending - March 1971 (Col. 5)'].iloc[30:36].sum()
ctot4=df['No. of breakdowns during the year ending - March 1971 (Col. 5)'].iloc[36:42].sum()

dtot1=df['No. of breakdowns per lakh G. Kms. During the year ending - March 1971 (Col. 6)'].iloc[0:12].sum()
dtot2=df['No. of breakdowns per lakh G. Kms. During the year ending - March 1971 (Col. 6)'].iloc[12:30].sum()
dtot3=df['No. of breakdowns per lakh G. Kms. During the year ending - March 1971 (Col. 6)'].iloc[30:36].sum()
dtot4=df['No. of breakdowns per lakh G. Kms. During the year ending - March 1971 (Col. 6)'].iloc[36:42].sum()

data={
    'Sub-category Service-wise':["City Services"," District Services"," Hill Passenger Services","Goods Services"],    
    'Total Number of Accidents during the half year ending - march. 1971 Service wise':[atot1,atot2,atot3,atot4],
    'Accidents per lakh gross kilometres during the half year ending - March. 1971 Service wise':[btot1,btot2,btot3,btot4],
    'No. of breakdowns during the year ending - March 1971 Service-wise':[ctot1,ctot2,ctot3,ctot4],
    'No. of breakdowns per lakh G. Kms. During the year ending - March 1971 Service-wise':[dtot1,dtot2,dtot3,dtot4]
        }
dtf=pd.DataFrame(data,columns=["Sub-category Service-wise","Total Number of Accidents during the half year ending - march. 1971 Service wise","Accidents per lakh gross kilometres during the half year ending - March. 1971 Service wise","No. of breakdowns during the year ending - March 1971 Service-wise","No. of breakdowns per lakh G. Kms. During the year ending - March 1971 Service-wise"])
dtf.set_index("Sub-category Service-wise",inplace=True)

#Renaming data for proper inference
dtf=dtf.rename(columns={"No. of breakdowns during the year ending - March 1971 Service-wise":"No. of breakdowns during the year ending - March 1971 Service-wise on a scale 0f 10000"})
dtf
```

```python
#Normalizing data for better visualization
dtf['No. of breakdowns during the year ending - March 1971 Service-wise on a scale 0f 10000']=dtf['No. of breakdowns during the year ending - March 1971 Service-wise on a scale 0f 10000']/100

#Plotting a suitable graph according to data
dtf.plot.bar(stacked=False,figsize=(40,20))
plt.show()
```
