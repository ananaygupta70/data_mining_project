```python
import pandas as pd
import matplotlib.pyplot as plt
from itertools import combinations
import numpy as np
df=pd.read_csv("C:\\dataMining\\8.4_Total_Number_of_Accidents_involving_injuries_to_persons_and_casualities.csv")
#Replacing all the non-numerical values with NA(Data Cleaning)
df=df.replace(to_replace ="@",value ="NaN")
df=df.replace(to_replace ="*",value ="NaN")
#Removing rows and columns with excess NA(Data Reduction)
df=df.dropna(thresh=len(df)-9,axis=1)
df=df.dropna(axis=0, how='any', thresh=4, subset=None, inplace=False)
#Filling NA with mean(Data cleaning)
df=df.fillna(df.mean(),inplace=False)
dff=df.copy()
df
```

```python
#Data Classification:
del df['Total No. of accidents involving injuries to persons (Col.3)']
df
```

```python
df=df.rename(columns = {"No. injured and killed in road accidents/ casualities - Injured (Col.4)":"Ratio of people injured"})
df=df.rename(columns = {"No. injured and killed in road accidents/ casualities - Killed (Col.5)":"Ratio of people killed"})
df=df.rename(columns = {"Country (Col.1)":"Country"})
df=df.rename(columns = {"Year (Col.2)":"Year"})
df
```

```python
df['Ratio of people injured']=df['Ratio of people injured']/df['No. injured and killed in road accidents/ casualities - Total (Col.6)']
df['Ratio of people killed']=df['Ratio of people killed']/df['No. injured and killed in road accidents/ casualities - Total (Col.6)']
del df['No. injured and killed in road accidents/ casualities - Total (Col.6)']
df
```

```python
df['Category']=""
# 1 means among injured in accidents, less than 5% People are killed in accidents
# 0 means among injured in accidents, more than 5% People are killed in accidents
# This reflects the health services of the country
df['Category']=np.where(df['Ratio of people killed']<0.05,1,0)
df
```

```python
# K-Means clustering starts
df=dff
df68=df[df['Year (Col.2)']==1968]
df69=df[df['Year (Col.2)']==1969]
df70=df[df['Year (Col.2)']==1970]
dff
```

```python
#storing number of accidents and persons killed in 1968
x=[]
y=[]
for i in range(0,len(df68)):
    x.append(df['Total No. of accidents involving injuries to persons (Col.3)'].iloc[i])
    y.append(df['No. injured and killed in road accidents/ casualities - Killed (Col.5)'].iloc[i])
```

```python
#storing indices for making combinations
ind=[]
for i in range(0,len(df68)):
    ind.append(i)
```

```python
#calculating minimum sse for each number of clusters
elb=[]
numclus=[]
for numofclus in range(1,12):
    comb=sum([list(map(list, combinations(ind, i))) for i in range(numofclus,numofclus+1)], [])
    minsse=1000000000000
    for j in range(1,len(comb)):
        cenx=[]
        ceny=[]
        for k in range(0,len(comb[j])):
            cenx.append(x[comb[j][k]])
            ceny.append(y[comb[j][k]])
        totsse=0
        for k in range(0,len(x)):
            mindis=1000000000000
            for l in range(0,len(cenx)):
                dis=(x[k]-cenx[l])*(x[k]-cenx[l]) + (y[k]-ceny[l])*(y[k]-ceny[l])
                if dis<mindis:
                    mindis=dis
            totsse = totsse + mindis
        if totsse<minsse:
            minsse=totsse
    elb.append(minsse)
    numclus.append(numofclus)
elb
```

```python
#plotting it in order to calculate the optimum number of clusters via elbow method

plt.plot(numclus, elb) 
  
# naming the x axis 
plt.xlabel('Number of Clusters') 
# naming the y axis 
plt.ylabel('Minimum of SSE') 
  
# giving a title to my graph 
plt.title('Calculating number of clusters using elbow method') 
  
# function to show the plot 
plt.show()
```

```python
#calculating the clusters with minimum sse for given number of clusters
minnsse=10000000000000
color1=[]
color2=[]
for i in range(0,len(x)):
    color1.append(-1)
    color2.append(-1)
#generating all possible combinations
comb1=sum([list(map(list, combinations(ind, i))) for i in range(3,4)], [])
for rep in range(0,len(comb1)):
    cenx=[]
    ceny=[]
    #appending corresponding points to indices
    for i in range(0,len(comb1[rep])):
        cenx.append(x[comb1[rep][i]])
        ceny.append(y[comb1[rep][i]])
    #performing k-means clustering
    for i in range(0,100):
        sse=0
        for j in range(0,len(x)):
            minn=10000000000000
            minin=-1
            for k in range(0,3):
                if((x[j]-cenx[k])*(x[j]-cenx[k])+(y[j]-ceny[k])*(y[j]-ceny[k])<minn):
                    minn=(x[j]-cenx[k])*(x[j]-cenx[k])+(y[j]-ceny[k])*(y[j]-ceny[k])
                    minin=k
            color2[j]=minin
            sse=sse+minn
        if(sse<minnsse):
            minnsse=sse
            for j in range(0,len(x)):
                #updating the color in case sse is less
                color1[j]=color2[j]
        for j in range(0,3):
            #calculating new centroids for each cluster
            newx=0
            newy=0
            xno=0
            yno=0
            for k in range(0,len(x)):
                if(color2[k]==j):
                    newx=newx+x[k]
                    newy=newy+y[k]
                    xno=xno+1
                    yno=yno+1
            cenx[j]=newx/xno
            ceny[j]=newy/yno
color1
```

```python
#poltting the obtained clusters on graph
a=np.array([x,y])
categories=np.array(color1)
colormap=np.array(['r','g','b'])
plt.rcParams["figure.figsize"] = (8, 8)
plt.scatter(a[0], a[1], s=20, c=colormap[categories])
plt.xlabel('Number of accidents')
plt.ylabel('Number of deaths')
plt.title('Clustering countries on the basis of number of road accidents and deaths in 1968')
plt.savefig('ScatterClassPlot.png')
plt.show()
```
